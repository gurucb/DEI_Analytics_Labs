```python
from pyspark.sql.functions import lit, udf, col, split
from pyspark.sql import functions as F
import pyspark
```

```python
authors = spark.read.csv(
    '/Volumes/pyspark_lessions/files/sample_files/library/authors.csv',
    inferSchema=True,
    header=True
)
authors_pd = authors.toPandas()
display(authors_pd.head())
display(authors_pd.tail())
```

```python
authors = authors.withColumn('FIRST_NAME', col=split(col('AUTHOR_NAME'), ' ')[0])
authors = authors.withColumn('Age',col=lit(30)+col('AUTHOR_ID'))
display(authors)
```

```python
books = spark.read.csv(
'/Volumes/pyspark_lessions/files/sample_files/library/books.csv',
inferSchema=True,
header=True
)
books_pd = books.toPandas()
books_pd.head()
```

```python
books = books.withColumn("BOOK_COST", F.col("BOOK_COST").cast("double"))
books = books.withColumn("BOOK_DISCOUNT",F.col("BOOK_DISCOUNT").cast("double"))
books.printSchema()
```
