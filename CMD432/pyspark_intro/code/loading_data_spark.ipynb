### Reading CSV Files

```python
authors = spark.read.csv(
    '/Volumes/pyspark_lessions/files/sample_files/library/authors.csv',
    inferSchema=True,
    header=True
)
authors_pd = authors.toPandas()
display(authors_pd.head())
display(authors_pd.tail())
```

```python
books = spark.read.csv(
'/Volumes/pyspark_lessions/files/sample_files/library/books.csv',
inferSchema=True,
header=True
)
books_pd = books.toPandas()
display(books_pd.head())
display(books_pd.tail())
```

```python
author_book = spark.read.csv(
    '/Volumes/pyspark_lessions/files/sample_files/library/book_author.csv',
    inferSchema=True,
    header=True
) 
author_book_pd = author_book.toPandas()
display(author_book_pd.head())
display(author_book_pd.tail())
```
