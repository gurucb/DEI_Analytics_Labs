```python
import seaborn as sns
from matplotlib import pyplot as plt
from pyspark.sql.functions import avg,round
import numpy as np
```

```python
raw_movie_data = spark.read.csv('/Volumes/movie_lens/raw_data/files/user_movie_ratings.csv',sep=',',header=True)
```

```python
display(raw_movie_data)
```

```python
raw_movie_data.count()
```

```python
ratings = raw_movie_data.select('Rating')
result = ratings.groupBy('Rating').count()
display(result)
```

```python

display((raw_movie_data.groupBy("User_ID").count()).agg(round(avg("count"))))
```

```python
result_pandas = result.sort("Rating").toPandas()
sns.barplot(x='Rating',y='count',data=result_pandas)
plt.show()
```

```python
item_results = (
                raw_movie_data.filter('Item_Id in ("931","200","265","100","101")')
                .groupBy("Item_Id")
                )
pivoted_item_result = item_results.pivot("Rating").count()
final_result = pivoted_item_result.na.fill(0)
```

```python
final_result.show()
```

```python
pandas_final_df = final_result.toPandas()
pandas_final_df.set_index('Item_Id',inplace=True)
print(pandas_final_df)
```

```python
ax = pandas_final_df.plot.bar(stacked=True, rot=0)
plt.show()
```
